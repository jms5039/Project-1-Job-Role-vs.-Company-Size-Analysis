```python
import pandas as pd
from pathlib import Path
```

```python
raw_data = Path("Resources/salaries.csv")
raw_data_salaries = pd.read_csv(raw_data)

raw_data_salaries.head(10)
```

# Filtering to employees who reside in the U.S.

```python
us_employee_salaries = raw_data_salaries.loc[raw_data_salaries["employee_residence"] == "US"]
us_employee_salaries
```

# Check for missing values

```python
us_employee_salaries.count()
```

# Filtering to 2023

```python
us_employee_2023_salaries = us_employee_salaries.loc[us_employee_salaries["work_year"] == 2023]
us_employee_2023_salaries
```

```python
us_employee_2023_salaries.count()
```

# Save as clean data file

```python
us_employee_2023_salaries.to_csv("Resources/clean_data.csv", index=False)
```
